```python
plott=store[(store["Sub-Category"]=='Phones') & (store["Region"]=='South') & (store['Sales'] > 1000)]
```

```python
plt.figure(figsize=(8,5))
plott.groupby("State")["Sales"].sum().plot(kind="bar")
plt.title("Phone Sales in  South Region")
plt.ylabel("Sales")
plt.xlabel("Sales",loc="center")
plt.xticks(rotation=20, ha='left')
plt.show()
```

```python
grp_nr.agg(["min",'max','mean'])
```

```python
sns.regplot(data=store,x= 'Sales',y="Profit",color='red')
plt.show()
```

```python
grouped = store.groupby('Region', as_index=False)['Sales'].sum()
grouped.plot(x = 'Region', y = 'Sales', label = 'Title', figsize = (7, 5), linewidth = 3,color = "red")
plt.ylabel('sales')
plt.title('My first plotting exercise')
plt.legend(loc = 'upper right')
plt.grid()
```

```python
segment_sales = store.groupby('Segment')['Sales'].sum()
plt.figure(figsize=(65,5))
plt.pie(segment_sales, labels=segment_sales.index, autopct='%1.1f%%', explode=(0,0,0.2),wedgeprops={'edgecolor': 'black', 'linewidth': 1}
,textprops={'fontsize':10,'color':'black','fontweight':'bold'},)
plt.title('Sales Share by Segment')
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(store['Discount'], store['Profit'], alpha=0.5)
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Impact of Discount on Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
monthly_sales = store.resample('M', on='Order Date')['Sales'].sum()
monthly_sales.plot(figsize=(7,4), linewidth=2)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid()
plt.tight_layout()
plt.show()
```
